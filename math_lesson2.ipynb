```python
import numpy as np

def matrix_addition(A, B):
    """
    Performs addition of two matrices
    
    Parameters:
    A (list of lists or numpy.ndarray): The first matrix
    B (list of lists or numpy.ndarray): The second matrix
    
    Returns:
    numpy.ndarray: The result of addition A + B
    """
    # Convert to numpy arrays
    A = np.array(A)
    B = np.array(B)
    
    # Verify that matrices are the same size
    if A.shape != B.shape:
        raise ValueError("Matrices must be the same size for addition")
    
    # Perform addition
    return A + B  # __add__

def matrix_subtraction(A, B):
    """
    Performs subtraction of two matrices
    
    Parameters:
    A (list of lists or numpy.ndarray): The first matrix
    B (list of lists or numpy.ndarray): The second matrix
    
    Returns:
    numpy.ndarray: The result of subtraction A - B
    """
    # Convert to numpy arrays
    A = np.array(A)
    B = np.array(B)
    
    # Verify that matrices are the same size
    if A.shape != B.shape:
        raise ValueError("Matrices must be the same size for subtraction")
    
    # Perform subtraction
    return A - B

def matrix_multiplication(A, B):
    """
    Performs multiplication of two matrices
    
    Parameters:
    A (list of lists or numpy.ndarray): The first matrix
    B (list of lists or numpy.ndarray): The second matrix
    
    Returns:
    numpy.ndarray: The result of multiplication A @ B
    """
    # Convert to numpy arrays
    A = np.array(A)
    B = np.array(B)
    
    # Verify that the number of columns in A equals the number of rows in B
    if A.shape[1] != B.shape[0]:
        raise ValueError("The number of columns in the first matrix must equal the number of rows in the second matrix")
    
    # Perform multiplication
    return A @ B  # or np.matmul(A, B)

# Define example matrices
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]
print("Matrix A:")
for row in A:
    print(row)

print("\nMatrix B:")
for row in B:
    print(row)

print()
# Addition
result_addition = matrix_addition(A, B)
print("Matrix addition:")
print(result_addition)

# Subtraction
result_subtraction = matrix_subtraction(A, B)
print("\nMatrix subtraction:")
print(result_subtraction)

# Multiplication
result_multiplication = matrix_multiplication(A, B)
print("\nMatrix multiplication:")
print(result_multiplication)
```

```python
import pandas as pd
import numpy as np  # Still need numpy for some operations

def matrix_addition_pandas(A, B):
   """
   Performs addition of two matrices using pandas
   
   Parameters:
   A (list of lists, numpy.ndarray, or pandas.DataFrame): The first matrix
   B (list of lists, numpy.ndarray, or pandas.DataFrame): The second matrix
   
   Returns:
   pandas.DataFrame: The result of addition A + B
   """
   # Convert to DataFrame
   if not isinstance(A, pd.DataFrame):
       A = pd.DataFrame(A)
   if not isinstance(B, pd.DataFrame):
       B = pd.DataFrame(B)
   
   # Verify that matrices are the same size
   if A.shape != B.shape:  # (3, 2) == (3, 2) a= A() a2 = A()  a == a2 ? No
       raise ValueError("Matrices must be the same size for addition")
   
   # Perform addition
   return A.add(B)  # or simply A + B

def matrix_subtraction_pandas(A, B):
   """
   Performs subtraction of two matrices using pandas
   
   Parameters:
   A (list of lists, numpy.ndarray, or pandas.DataFrame): The first matrix
   B (list of lists, numpy.ndarray, or pandas.DataFrame): The second matrix
   
   Returns:
   pandas.DataFrame: The result of subtraction A - B
   """
   # Convert to DataFrame
   if not isinstance(A, pd.DataFrame):
       A = pd.DataFrame(A)
   if not isinstance(B, pd.DataFrame):
       B = pd.DataFrame(B)
   
   # Verify that matrices are the same size
   if A.shape != B.shape:
       raise ValueError("Matrices must be the same size for subtraction")
   
   # Perform subtraction
   return A.sub(B)  # or simply A - B

def matrix_multiplication_pandas(A, B):
   """
   Performs multiplication of two matrices using pandas
   
   Parameters:
   A (list of lists, numpy.ndarray, or pandas.DataFrame): The first matrix
   B (list of lists, numpy.ndarray, or pandas.DataFrame): The second matrix
   
   Returns:
   pandas.DataFrame: The result of multiplication A @ B
   """
   # Convert to DataFrame
   if not isinstance(A, pd.DataFrame):
       A = pd.DataFrame(A)
   if not isinstance(B, pd.DataFrame):
       B = pd.DataFrame(B)
   
   # Verify that the number of columns in A equals the number of rows in B
   if A.shape[1] != B.shape[0]:
       raise ValueError("The number of columns in the first matrix must equal the number of rows in the second matrix")
   
   # Perform matrix multiplication
   return pd.DataFrame(np.dot(A, B))  # or A.dot(B)

# Define example matrices
A = pd.DataFrame([[1, 2], [3, 4]])
B = pd.DataFrame([[5, 6], [7, 8]])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

# Addition
result_addition = matrix_addition_pandas(A, B)
print("\nMatrix addition:")
print(result_addition)

# Subtraction
result_subtraction = matrix_subtraction_pandas(A, B)
print("\nMatrix subtraction:")
print(result_subtraction)

# Multiplication
result_multiplication = matrix_multiplication_pandas(A, B)
print("\nMatrix multiplication:")
print(result_multiplication)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_vector(vector, color='blue', ax=None, label=None):
    """
    Draws a vector from the origin with the given color
    
    Parameters:
    vector : list or numpy array of size 2 or 3 representing the vector
    color : string representing the color (default: 'blue')
    ax : axis object for plotting (optional)
    label : label for the vector (optional)
    
    Returns:
    ax : the axis object where the vector was drawn
    """
    vector = np.array(vector)
    
    # Check if it's a 2D or 3D vector
    if len(vector) == 2:
        # 2D vector
        if ax is None:
            fig, ax = plt.subplots(figsize=(8, 8))
        
        # Draw the vector from the origin
        ax.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color=color, label=label)
        
        # Set axis limits
        max_val = max(abs(vector[0]), abs(vector[1])) * 1.2
        ax.set_xlim([-max_val, max_val])
        ax.set_ylim([-max_val, max_val])
        
        # Add grid, axes, and titles
        ax.grid(True)
        ax.axhline(y=0, color='k', linestyle='-', alpha=0.3)
        ax.axvline(x=0, color='k', linestyle='-', alpha=0.3)
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_title('2D Vector Plot')
        
    elif len(vector) == 3:
        # 3D vector
        if ax is None:
            fig = plt.figure(figsize=(10, 8))
            ax = fig.add_subplot(111, projection='3d')
        
        # Draw the vector from the origin
        ax.quiver(0, 0, 0, vector[0], vector[1], vector[2], color=color, label=label)
        
        # Set axis limits
        max_val = max(abs(vector[0]), abs(vector[1]), abs(vector[2])) * 1.2
        ax.set_xlim([0, max_val])
        ax.set_ylim([0, max_val])
        ax.set_zlim([0, max_val])
        
        # Add titles
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Z')
        ax.set_title('3D Vector Plot')
    
    # Add legend if label is provided
    if label:
        ax.legend()
    
    return ax

# Example 1: Single 2D vector
v1 = [30, 4]
draw_vector(v1, color='red', label='Vector 1')
plt.show()

```

```python
fig, ax = plt.subplots(figsize=(8, 8))
draw_vector([3, 4], color='red', ax=ax, label='Vector 1')
draw_vector([2, -3], color='blue', ax=ax, label='Vector 2')
draw_vector([-4, 1], color='green', ax=ax, label='Vector 3')
plt.title('Multiple 2D Vectors')
plt.legend()
plt.show()
```

```python
# Example 3: 3D vector
v3d = [2, 3, 4]
draw_vector(v3d, color='purple', label='3D Vector')
plt.show()
```
